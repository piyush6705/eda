```python
numeric_cols = df.select_dtypes(include= ['int','float']).columns
corr_matrix =df[numeric_cols].corr()
print(corr_matrix)
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(corr_matrix, annot =True, cmap ='coolwarm',linewidth =5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
position_counts = df['Position'].value_counts()
position_counts
```

```python
plt.pie(position_counts,labels=position_counts.index, autopct = "%1.1f%%", startangle=90)
plt.title("Distribution of Positions")
plt.show()
```

```python
avg_attack_by_country =df.groupby('Country')['Attack'].mean()
```

```python
avg_attack_by_country.sort_values(ascending =False)
```

```python
avg_attack_by_country.sort_values(ascending=False).head(4).plot(kind='bar')
plt.title('AVG attck by country')
plt.xlabel('Country')
plt.ylabel('Average Attack ')
plt.show()
```

```python
avg_serve_by_age = df.groupby('Age')['Serve'].mean()
```

```python
plt.scatter(df['Block'], df['Receive'])
plt.title("Scatter Plot: Block vs Receive")
plt.xlabel("block")
plt.ylabel('Receive')
plt.show()
```

```python
sns.boxplot(x=df['Serve'])
plt.title("Box PLot: Distribution of Serve Values")
plt.xlabel("Serve")
plt.show()
```

```python
plt.hist(df["Age"],bins=20, color='skyblue',edgecolor='black')
plt.title("Distribution of Age")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
avg_attack_by_position =df.groupby('Position')['Attack'].mean()
```

```python
avg_attack_by_position.sort_values(ascending=True).plot(kind='bar',color='green')
plt.title('Average Attack by position')
plt.xlabel('position')
plt.ylabel('Average Attack')
plt.show()
```

```python
serve_trend_by_age =df.groupby('Age')['Serve'].mean()
```

```python
serve_trend_by_age
```

```python
serve_trend_by_age.plot(kind='line',marker= 'o', linestyle ='-', color ='orange')
plt.title('Serve Trends Over age Groups')
plt.xlabel('Age')
plt.ylabel('Average Serve')
plt.show()
```

```python
total_attack_block_by_county =df.groupby('Country')[['Attack','Block']].sum()
```

```python
total_attack_block_by_county
```

```python
total_attack_block_by_county.sort_values(ascending=False ,by='Attack' ).plot(kind='bar',stacked=True, colormap ='viridis')
plt.title('Total Attack and Block by County')
plt.xlabel('Country')
plt.ylabel('Total Value')
plt.show()
```
