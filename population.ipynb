```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("population.csv")
```

```python
df
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
sns.histplot(df['2023 population'], bins =10, kde =True)
plt.title("DIstribution of 2023 Population")
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

```python
df.groupby('continent')['2023 population'].mean().sort_values(ascending =False).head(2)
```

```python
top_10=df.sort_values(by ='2023 population', ascending =False).head(10)
```

```python
top_10
```

```python
sns.barplot(x='2023 population', y ='country', data =top_10)
plt.title("Top 10 countries in 2023")
plt.xlabel("Population")
plt.ylabel("country")
plt.show()
```

```python
sns.scatterplot(x='2023 population', y='density (km²)', data=df)
plt.title("population and density scatter plot")
plt.xlabel("Dentisy")
plt.ylabel("Density")
plt.show()
```

```python
df.groupby('continent')['area (km²)'].sum().sort_values(ascending=False)
```

```python
sns.violinplot(x='continent', y='density (km²)', data=df, palette ='viridis')
plt.title('Population Density Distribution Across Continents')
plt.xlabel("Continent")
plt.ylabel("Population density")
plt.show()
```

```python
sns.boxplot(x='continent', y='density (km²)', data=df, palette ='muted',showfliers=False)
plt.title('Population Density Distribution Across Continents')
plt.xlabel("Continent")
plt.ylabel("Population density")
plt.show()
```
